```python
df=[]
```

```python
for i in range(1,5):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key={api}&language=en-US&page={i}')
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'release_date','vote_average', 'vote_count','overview']]
    df.append(temp_df)
```

```python
df = pd.concat(df, ignore_index=True)
```
